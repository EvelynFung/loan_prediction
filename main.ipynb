Data Cleaning

```python
df.dropna(subset=['BORROWER_NAME', 'CITY', 'STATE', 'BANK', 'BANK_STATE', 'IS_NEW', 'IS_REVOLVER', 'IS_LOW_DOC', 'DISBURSEMENT_DATE', 'LOAN_STATUS'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.drop(columns=['BORROWER_NAME', 'CITY', 'ZIP', 'BANK'], inplace=True)
```

```python
# Converting the entries with dollars to floats
df[['DISBURSEMENT_AMOUNT', 'BALANCE_AMOUNT', 'CHARGE_OFF_AMOUNT', 'LOAN_AMOUNT', 'SBA_APPROVED_AMOUNT']] = \
df[['DISBURSEMENT_AMOUNT', 'BALANCE_AMOUNT', 'CHARGE_OFF_AMOUNT', 'LOAN_AMOUNT', 'SBA_APPROVED_AMOUNT']].applymap(lambda x: x.strip().replace('$', '').replace(',', ''))
```

```python
df.APPROVAL_YEAR.unique() # Some year has an A to it so we want to remove that
```

```python
df['APPROVAL_YEAR'] = df['APPROVAL_YEAR'].apply(lambda x: int(x.replace('A', '')) if isinstance(x, str) else x)
```

```python
df = df.astype({'IS_NEW': 'int64', 'IS_URBAN': 'str', 'DISBURSEMENT_AMOUNT': 'float', 'BALANCE_AMOUNT': 'float',
                          'CHARGE_OFF_AMOUNT': 'float', 'LOAN_AMOUNT': 'float', 'SBA_APPROVED_AMOUNT': 'float'})
```

```python
# Moving onto the industry, we will first extract the first two digits
df['INDUSTRY'] = df['INDUSTRY_ID'].astype('str').apply(lambda x: x[:2])

# Map it to the corresponding industry
df['INDUSTRY'] = df['INDUSTRY'].map({
    '11': 'Ag/For/Fish/Hunt',
    '21': 'Min/Quar/Oil_Gas_ext',
    '22': 'Utilities',
    '23': 'Construction',
    '31': 'Manufacturing',
    '32': 'Manufacturing',
    '33': 'Manufacturing',
    '42': 'Wholesale_trade',
    '44': 'Retail_trade',
    '45': 'Retail_trade',
    '48': 'Trans/Ware',
    '49': 'Trans/Ware',
    '51': 'Information',
    '52': 'Finance/Insurance',
    '53': 'RE/Rental/Lease',
    '54': 'Prof/Science/Tech',
    '55': 'Mgmt_comp',
    '56': 'Admin_sup/Waste_Mgmt_Rem',
    '61': 'Educational',
    '62': 'Healthcare/Social_assist',
    '71': 'Arts/Entertain/Rec',
    '72': 'Accom/Food_serv',
    '81': 'Other_no_pub',
    '92': 'Public_Admin'
})

# Drop any null values
df.dropna(subset = ['INDUSTRY'], inplace = True)
df.drop(columns = ['INDUSTRY_ID'], inplace = True)
```

```python
df.FRANCHISE_CODE.unique()
```

```python
# Convert franchise code to indicate whether it is a franchise or not
df.loc[(df['FRANCHISE_CODE'] <= 1), 'IS_Franchise'] = 0
df.loc[(df['FRANCHISE_CODE'] > 1), 'IS_Franchise'] = 1
df.drop(columns = ['FRANCHISE_CODE'], inplace = True)
```

```python
# Convert is_new to a binary column
df = df[(df['IS_NEW'] == 1) | (df['IS_NEW'] == 2)]

df.loc[(df['IS_NEW'] == 1), 'NEW'] = 0
df.loc[(df['IS_NEW'] == 2), 'NEW'] = 1
df.drop(columns = ['IS_NEW'], inplace = True)
```

```python
df.IS_REVOLVER.unique()
```

```python
df.IS_LOW_DOC.unique()
```

```python
# Convert is_revolver and is_low_doc to a binary column
df = df[(df.IS_REVOLVER == 'Y') | (df.IS_REVOLVER == 'N')]
df = df[(df.IS_LOW_DOC == 'Y') | (df.IS_LOW_DOC == 'N')]

df['IS_REVOLVER'] = np.where(df['IS_REVOLVER'] == 'N', 0, 1)
df['IS_LOW_DOC'] = np.where(df['IS_LOW_DOC'] == 'N', 0, 1)
```

```python
df.IS_REVOLVER.unique()
```

```python
df.IS_LOW_DOC.unique()
```

```python
df[['APPROVAL_DATE', 'DISBURSEMENT_DATE']] = df[['APPROVAL_DATE', 'DISBURSEMENT_DATE']].apply(pd.to_datetime)
```

```python
df['DAYS_TO_DIBURSEMENT'] = df['DISBURSEMENT_DATE'] - df['APPROVAL_DATE']
```

```python
df['DAYS_TO_DIBURSEMENT']
```

```python
# Remove the days at the end of the days_to_disbursement
df['DAYS_TO_DIBURSEMENT'] = df['DAYS_TO_DIBURSEMENT'].astype('str').apply(lambda x: x[:x.index('d') - 1]).astype('int64')
```

```python
df[df['DAYS_TO_DIBURSEMENT'] <= 0]
```

```python
df = df[df['DAYS_TO_DIBURSEMENT'] >= 0]
df
```

```python
df.LOAN_STATUS.unique()
```

```python
# Convert the loan status to a binary column
df['DEFAULT'] = np.where(df['LOAN_STATUS'] == 'P I F', 0, 1)
df.drop(columns = ['LOAN_STATUS'], inplace = True)
```

```python
df['DisbursementYEAR'] = df['DISBURSEMENT_DATE'].map(lambda x: x.year)
```

```python
df['DisbursementYEAR']
```

```python
df['SAME_STATE'] = np.where(df['STATE'] == df['BANK_STATE'], 1, 0)
```

```python
df['SBA_APPROVED_PERCENTAGE'] = df['SBA_APPROVED_AMOUNT'] / df['LOAN_AMOUNT']
```

```python
df = df.astype({'IS_Franchise': 'int64', 'NEW': 'int64'})
```

```python
df.drop(columns=['APPROVAL_DATE',
                      'DEFAULT_DATE', 'DISBURSEMENT_DATE', 'BALANCE_AMOUNT', 'CHARGE_OFF_AMOUNT', 'SBA_APPROVED_AMOUNT'], inplace=True)
```

```python
df.isnull().sum()
```

```python
# when the term is longer than 20 years which is 240 months we can assume that the loan is for real estate
df['REAL_ESTATE'] = np.where(df['TERM'] >= 240, 1, 0)
```

```python
df['DISBURSED_GREATER_THAN_LOAN'] = np.where(df['DISBURSEMENT_AMOUNT'] > df['LOAN_AMOUNT'], 1, 0)
```

```python
df.dropna(subset=['TREASURY_YIELD', 'CPI_INDEX', 'GDP', 'MORTGAGE_30_US_FIXED', 'UNRATE', 'INDPRO_INDEX', 'UMCSENT_INDEX', 'CSUSHPINSA_INDEX', 'CP_INDEX', 'FEDFUNDS_RATE'], inplace=True)
```

```python
# Machine Learning
drive.mount('/content/gdrive')
df = pd.read_csv('gdrive/My Drive/df_new.csv')
df.head()
```

```python
df = df.drop(['Unnamed: 0'],axis=1)
```

```python
df = pd.get_dummies(df)

df.head()
```

```python
# Setting Features and Target
X = df.drop(['DEFAULT'],axis=1)
y = df['DEFAULT']

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest
from sklearn.linear_model import LogisticRegression
```

```python
# Standar Scaler for normal data
scale = StandardScaler()
X_scld = scale.fit_transform(X)
```

```python
# Train Test Split
X_train, X_val, y_train, y_val = train_test_split(X_scld, y, test_size=0.3)
```

```python
# Logistic Regression (Classification)
from sklearn.metrics import classification_report
lr = LogisticRegression(random_state = 42)

lr.fit(X_train, y_train)
y_pred = lr.predict(X_val)

print(classification_report(y_val, y_pred, digits = 3))
print("Testing accuracy is:",accuracy_score(y_val, y_pred))
```

```python
# Decision Tree (Classification)
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(random_state = 42, max_leaf_nodes=40)
model_dtc = dtc.fit(X_train, y_train)

y_pred = dtc.predict(X_val)

print(classification_report(y_val, y_pred, digits = 3))
print("Testing accuracy is:",accuracy_score(y_val, y_pred))
```

```python
cm = confusion_matrix(y_val, y_pred)
print(cm)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['No Default', 'Default'], yticklabels=['No Default', 'Default'])
plt.title('Confusion Matrix for Decision Tree')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
# Random Forest (Classification)
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state = 42, max_depth=20)
model_rfc = rfc.fit(X_train, y_train)

y_pred = rfc.predict(X_val)

print(classification_report(y_val, y_pred, digits = 3))
print("Testing accuracy is:",accuracy_score(y_val, y_pred))
```

```python
cm2 = confusion_matrix(y_val, y_pred)
```
